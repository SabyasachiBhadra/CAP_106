```python
# count the run length hours in each Run
df_run_hours = pd.DataFrame(df.groupby('RunNo.').count()['RunDurationHours'])
```

```python
# count number of days in each run
df_run_hours['RunDurationDays'] = (df_run_hours['RunDurationHours'] / 24)
```

```python
# store this run duration days info in a dataframe
df_run_hours.to_excel("C:/Users/SABYA/Aritar01/CAP_106/CAP_106_Research_Env/EDA_Report/historic_run_days.xlsx", index=False)
```

```python
# store the last RunDurationHours value from each runs
df_run_hours_last_value = pd.DataFrame(df.groupby("RunNo.").last()["RunDurationHours"])
```

```python
# store the last RunDurationHours value from each runs in a dictionary
last_val_from_each_run_hours = dict()
for hrs in df_run_hours_last_value["RunDurationHours"]:
    last_val_from_each_run_hours[df_run_hours_last_value[df_run_hours_last_value["RunDurationHours"]==hrs].index.values[0]] = hrs
```

```python
last_val_from_each_run_hours
```

```python
# introducing a relative run length column to the sample dataset
run_length_hours = []
for run_no, run_hours in zip(df["RunNo."], df["RunDurationHours"]):
    if run_no in df["RunNo."].unique().tolist():
        run_length_hours.append(last_val_from_each_run_hours[run_no] - run_hours)
# add this new relative run length column to the sample dataset
df["RelativeRunLength"] = run_length_hours 
```

```python
## How we can drop out highly correlated features in Python
def Snippet_127_Ex_2():
   
    print('**How to drop out highly correlated features in Python**\n')

    # loading libraries
    import pandas as pd
    import numpy as np
    from sklearn import datasets

    iris = datasets.load_iris()

    # Creating features and target
    X = iris.data
    y = iris.target
    
    # Converting feature matrix into DataFrame
    df = pd.DataFrame(X)

    # Viewing the data frame
    print(df.head())

    # Creating correlation matrix
    cor_matrix = df.corr().abs()
    print(); print(cor_matrix)

    # Selecting upper triangle of correlation matrix
    upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),
                                      k=1).astype(np.bool))
    print(); print(upper_tri)

    # Finding index of feature columns with correlation greater than 0.95
    to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
    print(); print(to_drop)

    # Droping Marked Features
    df1 = df.drop(df.columns[to_drop], axis=1)
    print(); print(df1.head())

Snippet_127_Ex_2()
```
